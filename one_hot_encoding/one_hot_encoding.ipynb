```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("carprices.csv")
data.head()
```

```python
%matplotlib inline
plt.xlabel("Mileage")
plt.ylabel("Sell Price($)")
plt.scatter("Mileage", "Sell Price($)", data=data)
plt.show()
```

#### One hot encoding

```python
dummies = pd.get_dummies(data["Car Model"])
dummies
```

```python
merge = pd.concat([data,dummies], axis= 1)
merge
```

```python
df = merge.drop(["Car Model", "Mercedez Benz C class"],axis = 1)
df
```

```python
# splitting the data in X and y
X = df.drop(["Sell Price($)"],axis = 1)
X
```

```python
y = df["Sell Price($)"]
y
```

```python
model = LinearRegression()
model.fit(X,y)
```

Model accuracy

```python
model.score(X,y)
```

Price of Audi A5 that is 4 yr old with mileage 45000

```python
model.predict([[45000,4,1,0]])
```

Price of BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1]])
```
